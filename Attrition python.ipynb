```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
#Age Distribution of Employeesplt.figure(figsize=(10, 6))
sns.histplot(Attrition['Age'], kde=True, bins=30)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(data=Attrition, x='Department', hue='Attrition',palette='pastel')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
#Attrition by Education
plt.figure(figsize=(10, 6))
sns.countplot(data=Attrition, x='Education', hue='Attrition')
plt.title('Attrition by Education')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
#Attrition by Marital Status
plt.figure(figsize=(10, 6))
sns.countplot(data=Attrition, x='MaritalStatus', hue='Attrition')
plt.title('Attrition by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```
